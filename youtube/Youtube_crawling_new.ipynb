```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import numpy as np
import pandas as pd
import bs4
import time
import json
import openpyxl
from selenium.webdriver.common.keys import Keys
```

```python
class Youtube_Crawler():
    def __init__(self,driver):
        self.driver = driver
        
    def move_to_channel(self, channel_name):
        self.driver.get('https://www.youtube.com/')
        time.sleep(0.6)
        
        search_block = self.driver.find_elements_by_xpath('//*[@id="search"]')
        search_block[0].send_keys(channel_name)
        click_button = self.driver.find_elements_by_css_selector('button#search-icon-legacy')
        click_button[0].click()
        time.sleep(1)
        
        channel_button = self.driver.find_elements_by_css_selector('a.channel-link')
        channel_button[1].click()
        time.sleep(0.6)
        
        video_button = self.driver.find_elements_by_xpath('//*[@id="tabsContent"]/tp-yt-paper-tab[2]')
        video_button[0].click()
        
        now_url = driver.current_url
        return now_url
    
    def get_videos_url(self, channel_name, PATH):
        video_url = list()
        
        url = self.move_to_channel(channel_name)
        
        self.driver.get(url)
        time.sleep(1)
        
        first_height=560
        driver.execute_script(f"window.scrollTo(0, {first_height});")
        while True:
            scroll_height = 2000
            document_height_before = driver.execute_script("return document.documentElement.scrollHeight")
            driver.execute_script(f"window.scrollTo(0, {document_height_before+scroll_height});")
            time.sleep(1.5)
            document_height_after = driver.execute_script("return document.documentElement.scrollHeight")
            if document_height_after == document_height_before:
                break
        
        video_id = driver.find_elements_by_css_selector('a#video-title')
        for i in video_id:
            video_url.append(i.get_attribute('href'))
        
        df = pd.DataFrame(video_url)
        
        df.to_csv(PATH+f'/{channel_name}_youtube_video_list.csv',index=None)
        
        return video_url
    
    def get_comments(self, url_list):
        comments=list()
        self.driver.implicitly_wait(5)
        
        for i in range(0,len(url_list)):
            self.driver.get(url_list[i])
            first_height=560
            driver.execute_script(f"window.scrollTo(0, {first_height});")
            while True:
                scroll_height = 2000
                document_height_before = driver.execute_script("return document.documentElement.scrollHeight")
                driver.execute_script(f"window.scrollTo(0, {document_height_before+scroll_height});")
                time.sleep(1.5)
                document_height_after = driver.execute_script("return document.documentElement.scrollHeight")
                if document_height_after == document_height_before:
                    break
            
            comments_sel = driver.find_elements_by_css_selector('yt-formatted-string#content-text')
            
            for j in range(0, len(comments_sel)):
                comments.append(comments_sel[j].text)
        
        return comments
```
