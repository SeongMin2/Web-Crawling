# 특정 게시판에서 댓글 모두 추출 (핵심)

#### 하지만 문제가 있음 15페이지가 초과되면 다음을 눌러서 다음 페이지로 가야하는데 
#### 여기서 다음 버튼을 누르게되면 현재 위치한 페이지+15의 페이지로 이동해버리게 됨
#### 따라서 15초과된 페이지가 있다면 15 이후 다시 1페이지로 돌아간 후 1페이지에서 다음을 누름으로써 16페이지로 이동시키게 함
#### 이렇게 다음 페이지로 가야할 수를 move_to_next 라고 저장하고 그 수에 따라서 위의 알고리즘으로 수행하도록 시킴
#### abs(move_to_next-n) >=2 이어야만 첫째 페이지로 이동시키고 '다음' 버튼 누르도록 함
#### 첫째 페이지란 1 ~ 15에서는 1, 16 ~ 30 에서는 16 


```python
def get_commentss2(numbers,whoareU,fileNameNum):
    comments=list()
    
    
    for i in range (0,len(numbers)):
        totalNum=0
        a=numbers[i]
        if(a=='16215'):
            continue
        url=f'https://gall.dcinside.com/board/view/?id=hit&no={numbers[i]}&page=1'
        driver.get(url)
    
        comment_num=driver.find_elements_by_id('comment_total_'+numbers[i])[0].text
        totalNum=int(comment_num)
        if totalNum == 0:
            continue
        elif totalNum/100 > 0 and totalNum%100 > 0:   
            hasPage=int(totalNum/100) + 1
        elif totalNum%100 == 0:  
            hasPage = int(totalNum/100)
        else:
            hasPage=1
        print("페이지 수 :"+str(hasPage))
        move_to_next=0
        if hasPage%15==0 and hasPage/15>1:
            move_to_next=int(hasPage/15 -1)
        elif hasPage==15:
            move_to_next=0
        else:
            move_to_next=int(hasPage/15)
        accessPage=0
        for n in range (0,move_to_next+1):
            # 우선 들어가자마자 1페이지에 대해서 크롤링 하기
            comment = driver.find_elements_by_css_selector('p.usertxt')
            for k in range (0,len(comment)):
                if comment[k].text == '-------------------------- 원본 댓글 -----------------------------' or comment[k].text == '해당 댓글은 삭제되었습니다.' or comment[k].text == '순위권, 반복리플, 욕설, 도배리플 등은 삭제됩니다.' or comment[k].text == '힛갤에 등록된 게시자에게 기념품을 드립니다. 연관갤러리 클릭하셔서 배송지를 남겨주세요!':
                    continue
                print(comment[k].text)
                comments.append(comment[k].text)
            accessPage=accessPage+1
            print("#########################################################################################################"+str(accessPage))
            # 이 후 만약 2페이지 이상 존재 시  
            if(hasPage>15):
                viewHasPage = 15
            else:
                viewHasPage = hasPage
            for m in range (0,viewHasPage-1):
                
                b=driver.find_elements_by_css_selector('div.cmt_paging>a')
                if(n>0):
                    b[m+2].send_keys(Keys.ENTER)   # .click으로 하게 되면 맨처음에 실행시켰을 때 ElementClickInterceptedException가 지속적으로 발생
                    accessPage=accessPage+1
                else:
                    b[m].send_keys(Keys.ENTER)
                    accessPage=accessPage+1
                comment = driver.find_elements_by_css_selector('p.usertxt')
                
                for k in range (0,len(comment)):
                    if comment[k].text == '-------------------------- 원본 댓글 -----------------------------' or comment[k].text == '해당 댓글은 삭제되었습니다.' or comment[k].text == '순위권, 반복리플, 욕설, 도배리플 등은 삭제됩니다.' or comment[k].text == '힛갤에 등록된 게시자에게 기념품을 드립니다. 연관갤러리 클릭하셔서 배송지를 남겨주세요!':
                        continue
                    print(comment[k].text)
                    comments.append(comment[k].text)
                print("#########################################################################################################"+str(accessPage))
            if(abs(n-(move_to_next+1))>=2):
                if(n==0):
                    driver.find_elements_by_css_selector('div.cmt_paging>a')[0].send_keys(Keys.ENTER)
                    driver.find_elements_by_css_selector('div.cmt_paging>a')[14].send_keys(Keys.ENTER)
                else:
                    driver.find_elements_by_css_selector('div.cmt_paging>a')[2].send_keys(Keys.ENTER)
                    driver.find_elements_by_css_selector('div.cmt_paging>a')[16].send_keys(Keys.ENTER)
                hasPage=hasPage-15
                
    if(len(comments)!=0):
        df = pandas.DataFrame(comments)
        if(whoareU==3):
            df.to_excel('dcinside_data/dcinside_result'+str(fileNameNum)+'('+str(numbers[0])+' to '+str(numbers[len(numbers)-1])+')'+'.xlsx',header=['comment'], index=None)
    else:
        print('추출한 기사에 댓글이 하나도 없습니다.')
        
    print('추출한 댓글 수 :',len(comments))
```

# Hit 갤러리에서 게시글 번호 추출(핵심)

```python
def getNumbers(url):
    postNumbers=list()
    driver.get(url)
    
    driver.find_elements_by_css_selector('div.bottom_paging_box>a')[15].click()
    hasPage = driver.find_elements_by_css_selector('div.bottom_paging_box>em')[0].text
    print("총 게시글 페이지 수(한 페이지 당 50개 게시글):",hasPage)
    driver.find_elements_by_css_selector('div.bottom_paging_box>a')[0].click()
    
    pageNum=int(input("추출할 게시물 페이지 수를 입력하세요:"))
    hasPage=pageNum
    
    move_to_next=0
    if(hasPage%15==0 and hasPage/15>1):
        move_to_next=hasPage/15 -1
    else:
        move_to_next=int(hasPage/15)
    
    for n in range (0,move_to_next+1):
        # 첫 페이지 먼저 글 번호 추출
        number = driver.find_elements_by_css_selector('td.gall_num')
        for i in range (0,len(number)):
            no = number[i].text
            if(no=='공지' or no=='설문'):
                continue
            postNumbers.append(no)
            
        if(hasPage>15):
                viewHasPage = 15
        else:
            viewHasPage = hasPage
        for m in range (0,viewHasPage-1):
            b=driver.find_elements_by_css_selector('div.bottom_paging_box>a')
            if(n>0):
                b[m+2].send_keys(Keys.ENTER)   # .click으로 하게 되면 맨처음에 실행시켰을 때 ElementClickInterceptedException가 지속적으로 발생
                    
            else:
                b[m].send_keys(Keys.ENTER)
                
            number = driver.find_elements_by_css_selector('td.gall_num')
            for i in range (0,len(number)):
                no = number[i].text
                if(no=='공지' or no=='설문'):
                    continue
                postNumbers.append(no)
        if(abs(n-(move_to_next+1))>=2):
            if(n==0):
                driver.find_elements_by_css_selector('div.bottom_paging_box>a')[0].send_keys(Keys.ENTER)
                driver.find_elements_by_css_selector('div.bottom_paging_box>a')[14].send_keys(Keys.ENTER)
            else:
                driver.find_elements_by_css_selector('div.bottom_paging_box>a')[2].send_keys(Keys.ENTER)
                driver.find_elements_by_css_selector('div.bottom_paging_box>a')[16].send_keys(Keys.ENTER)
            hasPage=hasPage-15
        
            
    return postNumbers
```

```python
url = 'https://gall.dcinside.com/board/lists?id=hit'
postNumbers=getNumbers(url)
print(postNumbers)
print("추출한 게시글 수:",len(postNumbers))
```
